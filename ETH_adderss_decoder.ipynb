```python
#split data
from keras import backend as K
dataset = df.values
np.random.shuffle(dataset)
```

```python
X=dataset[:,0:40]
Y=dataset[:,40:]


X_train1 = K.cast_to_floatx(X)
Y_train1 = K.cast_to_floatx(Y)

'''X=X.reshape(50000,input_dim).astype("int32")'''
```

```python
model=Sequential()
model.add(Dense(64,input_shape=(40,),activation="sigmoid"))
model.add(Dense(128,activation="sigmoid"))
model.add(Dense(256,activation="tanh"))
model.add(Dense(128,activation="sigmoid"))
model.add(Dense(64,activation="linear"))
model.summary()
#更換activation func
```

```python
model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(momentum=0.9), metrics=['mse','mae','RootMeanSquaredError'])
#更換loss func and optimizer
```

```python
model.fit(X_train1,Y_train1,epochs=15,batch_size=128)
```

```python
#loss,accuracy = model.evaluate(X_train1,Y_train1)
print(model.evaluate(X_train1,Y_train1))
#print(f"準確度:{accuracy:.2f}")
```
